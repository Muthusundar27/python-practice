4. Write a Python program to plot two or more lines on same plot with suitable legends of each line. 

```python
x=[10,20,30]
y=[20,40,10]
y1=[40,10,30]
plt.plot(x,y, label = 'line1')
plt.plot(x,y1,'g',label = 'line2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Two or more lines on same plot with suitable legends')
plt.legend()
```

5. Write a Python program to plot two or more lines with legends, different widths and colors. 

```python
x=[10,20,30]
y=[20,40,10]
y1=[40,10,30]
plt.plot(x,y,'b', label = 'line1-width-3',linewidth=3)
plt.plot(x,y1,'r', label = 'line2-width-5',linewidth=5)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Two or more lines on same plot with suitable legends')
plt.legend()
```

6. Write a Python program to plot two or more lines with different styles.

```python
x=[10,20,30]
y=[20,40,10]
y1=[40,10,30]
plt.plot(x,y,'b',label = 'line1-dotted',linestyle='dotted',linewidth=3)
plt.plot(x,y1,'r',label = 'line2-dashed',linestyle='dotted',linewidth=5)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Two or more lines on same plot with suitable legends')
plt.legend()
```

7. Write a Python program to plot two or more lines and set the line markers. 

```python
x=[1,4,5,6,7]
y=[2,6,3,6,3]
plt.plot(x,y,color='red', linestyle='dashdot', linewidth = 1, marker='o', markerfacecolor='blue', markersize=8)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Display marker')
```

 8. Write a Python program to display the current axis limits values and set new axis values.

```python
X = range(1, 50)
Y = [value * 3 for value in X]
plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Draw a line.')
plt.axis()
plt.axis([0, 100, 0, 200])
```
